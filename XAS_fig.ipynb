```python
import numpy as np
import os, json
import matplotlib.pyplot as plt
```

```python
sel_xanes_dic = {}
with open('xane.json', 'r') as f:
    sel_xanes_dic = json.load(f)
```

```python
fig, ax = plt.subplots(2,1,figsize=(4, 4), sharex=True)
pstr = ['n1_brrich_fresh', 'n1_brrich_aflight', 'n3_irich_aflight', 'n3_irich_fresh']
color = ['lightcoral', 'lightsteelblue', 'royalblue', 'orangered']
cts = 0
for p, c in zip(pstr, color):
    xval = sel_xanes_dic[p]['brag_eng']
    yval = np.sum(sel_xanes_dic[p]['Br_int_cts'], axis = 1) / sel_xanes_dic[p]['usic']
    normyval = yval / np.max(yval)
    ax[0].plot(xval, normyval, label=p, color=c)
    ax[1].plot(xval[1:], np.diff(normyval)+cts*0.05, label=p, color=c)
    cts += 1
ax[0].legend(frameon=False)
ax[1].set_xlim(left=13.46, right=13.49)
ax[0].set_xlim(left=13.46, right=13.49)

plt.tight_layout()
# plt.savefig('./cmp_n1n3_xanes.svg', dpi = 300)
```
